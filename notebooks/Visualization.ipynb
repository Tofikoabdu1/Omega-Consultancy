```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
IN = Path("../data/processed/all_reviews_with_themes.csv")
OUT = Path("../data/processed/figures")
OUT.mkdir(parents=True, exist_ok=True)
```

```python
df = pd.read_csv(IN)
```

```python
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='rating', hue='bank')
plt.title('Rating distribution by bank')
plt.savefig(OUT/'rating_distribution.png', bbox_inches='tight')
plt.show()
plt.close()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='sentiment_label', hue='bank')
plt.title('Sentiment counts by bank')
plt.savefig(OUT/'sentiment_counts.png', bbox_inches='tight')
plt.show()
plt.close()
```

```python
plt.figure(figsize=(10,6))
top_themes = df['themes'].value_counts().nlargest(15)
sns.barplot(y=top_themes.index, x=top_themes.values)
plt.title('Top themes (top 15)')
plt.savefig(OUT/'top_themes.png', bbox_inches='tight')
plt.show()
plt.close()
```

```python
print("Saved figures to", OUT)
```
